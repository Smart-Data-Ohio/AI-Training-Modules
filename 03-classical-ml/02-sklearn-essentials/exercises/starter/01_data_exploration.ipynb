## Part 2: Handle Missing Data

Now that you know where the gaps are, decide what to do about them.

**Options for each column with missing values:**
- **Drop rows** — if you have plenty of data and the missing values are random
- **Fill with mean/median** — good for numerical columns (median is more robust to outliers)
- **Fill with mode** — good for categorical columns (most frequent value)
- **Drop the column** — if too many values are missing to be useful

Think about which strategy makes sense for each column before coding.
