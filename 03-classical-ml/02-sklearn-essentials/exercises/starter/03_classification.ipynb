**Why does scaling matter for SVM but not Random Forest?**

SVMs compute distances between data points. If one feature ranges from 0-300 (cholesterol) and another from 0-1 (fasting blood sugar), cholesterol will dominate the distance calculation. Scaling puts everything on the same scale.

Random forests make decisions based on thresholds ("is cholesterol > 200?"), so the absolute scale doesn't matter.

**The key question:** If this were a medical screening tool, which metric would you prioritize — precision or recall? Why?

Think about it this way:
- **High precision, lower recall:** Fewer false alarms, but we might miss some sick patients
- **High recall, lower precision:** We catch most sick patients, but some healthy patients get flagged too

In medical screening, missing a sick patient (false negative) is usually worse than a false alarm (false positive), because the false positive just leads to more testing while the false negative means a missed diagnosis.
